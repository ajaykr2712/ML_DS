```python
import matplotlib.pyplot as plt
import numpy as np


years = list(range(2010, 2026))

contributions = [
    2,   # 2010
    203,   # 2011
    211,  # 2012
    341,  # 2013
    2763,  # 2014
    799,  # 2015
    648,  # 2016 
    2147,  # 2017
    4806,  # 2018 (peak year)
    3398,  # 2019
    3251,  # 2020
    1645,  # 2021
    2083,  # 2022
    1875,  # 2023
    2713,  # 2024
    +165  # 2025
]


plt.figure(figsize=(12, 6))

# Plot the data with a line and markers
plt.plot(years, contributions, marker='o', linestyle='-', linewidth=2, markersize=8)

# Customize the plot
plt.title('GitHub Contributions Over Years (2010-2025)', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Contributions', fontsize=12)


max_contribution = max(contributions)
plt.yticks(np.arange(0, max_contribution + 400, 200))


plt.grid(True, linestyle='--', alpha=0.7)


plt.xticks(years, rotation=45)


for i, value in enumerate(contributions):
    plt.text(years[i], value + 100, str(value), 
             ha='center', va='bottom')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
```
